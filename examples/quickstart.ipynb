# SynthHub Quickstart

This notebook uses the lightweight `independent` backend so it runs without optional research-engine dependencies.

```python
import pandas as pd
from synthhub import Synthesizer

real_df = pd.DataFrame({
    "age": [21, 34, 37, 45, 52, 23, 41, 29, 62, 31],
    "city": ["A", "B", "A", "C", "B", "A", "C", "C", "B", "A"],
    "churn": [0, 1, 0, 1, 1, 0, 1, 0, 1, 0],
})

synth = Synthesizer(method="independent", epsilon=1.0, random_state=0)
synth.fit(real_df)
synthetic_df = synth.sample(100)
synthetic_df.head()
```

```python
report = synth.evaluate(real_df, synthetic_df, target="churn")
report.to_dict()
```
